```python
train_df
```

```python
# create another dataframe for clutter data and filter for object_not_present_issue == True

clutter_df = train_df[train_df['object_not_present_issue'] == False]
# check that index contains the clutter string
clutter_df = clutter_df[clutter_df.index.str.contains('clutter')]
clutter_df.drop(columns=['object_not_present_issue', 'object_bounding_box', 'framing_issue', 'object_not_present_issue', 'framing_issue','viewpoint_issue','blur_issue','occlusion_issue','overexposed_issue','underexposed_issue'], inplace=True)

#reset index and set the index name to 'image_id'
clutter_df.reset_index(inplace=True)
clutter_df.rename(columns={'index':'path'}, inplace=True)

# split this string 'P573--orbit-reader-20-braille-display--clutter--fjhyeFn_6UJSO7bSCot0BOAsuw61L3K7lB-1-Hdlq8U-00001.jpg' into -- chunks
clutter_df['user'] = clutter_df['path'].apply(lambda x: x.split('--')[0])
clutter_df['object'] = clutter_df['path'].apply(lambda x: x.split('--')[1].replace('-', ' '))
clutter_df['type'] = clutter_df['path'].apply(lambda x: x.split('--')[2])
clutter_df['image_id'] = clutter_df['path'].apply(lambda x: x.split('--', maxsplit=3)[3])
clutter_df['image_id'] = clutter_df['image_id'].apply(lambda x: x.split('.')[0][:-6])
clutter_df['img_number'] = clutter_df['path'].apply(lambda x: x.split('.')[0][-5:])
clutter_df['folder'] = clutter_df['path'].apply(lambda x: x.split('--')[0] + '--' + x.split('--')[1] + '--' + x.split('--')[2] + '--' + x.split('--', maxsplit=3)[3].split('.')[0][:-6])
clutter_df['complete_path'] = '/orbit_benchmark_224/train/' + clutter_df['user'] + '/' + clutter_df['object'] + '/' + clutter_df['type'] + '/' + clutter_df['folder'] + '/' + clutter_df['path']   


clutter_df.drop(columns=['path'], inplace=True)


clutter_df

```

```python
# to save the dataframe to a parquet file
clutter_df.to_parquet('../../orbit_benchmark_224/parquet/train_clutter.parquet')
```
