```python
import os
import json
import pandas as pd
import glob
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
```

## For JSON

```python
# read names of json files in a list
json_files = glob.glob('/home/branislava.jankovic/Documents/object_detection/project/orbit-challenge-2024/dataset/annotation/orbit_extra_annotations/test/*.json')
print(len(json_files))
```

```python
# load files into new_data
new_data  = [pd.read_json(file).transpose() for file in json_files]
```

```python
# sampling
step = 3
range = 30
for data in new_data:
    name = data.index[0][:-10]
    if '--clean--' in name:
        data = data.iloc[range:-range]
        data = data.iloc[::step]
    data.to_json('/home/branislava.jankovic/Documents/object_detection/project/orbit-challenge-2024/dataset/annotation/orbit_extra_annotations/new_test_big/'+name+'.json', orient='index')
```

## For Images

```python
json_files = glob.glob('/home/branislava.jankovic/Documents/object_detection/project/orbit-challenge-2024/dataset/annotation/orbit_extra_annotations/new_test_big/*.json')
print(len(json_files))

# load files into new_data
new_data = [pd.read_json(file).transpose() for file in json_files]
```

```python
import matplotlib.pyplot as plt
for data in new_data:
    names = data.index
    for name in names:
        user = name.split("--")[0]
        object = name.split("--")[1].replace("-", " ")
        clean_or_clutter = name.split("--")[2]
        video = name[:-10]
        # print(user, object, clean_or_clutter, video)

        # read image
        image = imread('/home/branislava.jankovic/Documents/object_detection/project/orbit-challenge-2024/dataset/train/'+user+'/'+object+'/'+clean_or_clutter+'/'+video+'/'+name)
        # plt.imshow(image)

        # save image
        full_path = os.path.join('/home/branislava.jankovic/Documents/object_detection/project/orbit-challenge-2024/dataset/new_train1/', user, object, clean_or_clutter, video)
        os.makedirs(full_path, exist_ok=True)
        imsave(full_path+'/'+name, image)
```
